```python
import hashlib
from pathlib import Path

def sha256_file(path: Path) -> str:
    h = hashlib.sha256()
    with path.open("rb") as f:
        for chunk in iter(lambda: f.read(65536), b""):
            h.update(chunk)
    return h.hexdigest()

# In this inert fixture, expected hashes are recorded in expected/digest_manifest.json.
# input_manifest.sha256 is the digest of the input manifest file only.
# Payload artifact hashes are recorded separately under evidence_items[].sha256.
```
