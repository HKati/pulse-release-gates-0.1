# PULSE memory / trace dashboard v0 (demo)

This notebook is a small, developer-focused dashboard built on top of the
memory / trace summariser v0 artefacts. It is intended to explore how
paradox axes and EPF signals behave over time, not to change any gate
decisions.

The notebook reads JSON artefacts such as:

- `paradox_history_v0.json`
- `delta_log_v0.json`
- `paradox_resolution_v0.json`
- `paradox_resolution_dashboard_v0.json`
- `topology_dashboard_v0.json` (optional)

and uses the shared panel helpers in `_memory_trace_panels_v0_cells.py` to
render a few views over time:

- paradox history across runs,
- stability / risk zone trends,
- EPF signal trends (if available),
- resolution hints based on the paradox history,
- a weighted Paradox histogram by zone (v0).

---

### How to run

1. Make sure the artefacts listed above exist under your artefact root
   (by default: `../artifacts` relative to this notebook).
2. Open this notebook from the repository checkout:
   `PULSE_safe_pack_v0/examples/PULSE_memory_trace_dashboard_v0_demo.ipynb`.
3. Run the first code cell:

   ```python
   from _memory_trace_panels_v0_cells import run_all_panels
   run_all_panels(globals())
   ```

4. Scroll through the rendered panels and adjust filters in individual
   cells as needed.
